```python
'''
# SELECT *

SELECT * FROM table;


 
# SELECT n

SELECT name , height FROM table;
 
 
 
# COUNT

SELECT COUNT(*) from table;




# UNIQUE VALUES

SELECT DISTINCT height FROM table;

SELECT DISTINCT(height, weight) FROM table;



# WHERE CONDITION -- =,<,>,!=

SELECT * FROM table WHERE height =/</>/!= 150;




#AND/OR

SELECT * FROM table WHERE height = 150 AND WEIGHT = 70;



#IN

SELECT * FROM table WHERE height in (150, 160);



#NULL/NOTNULL

SELECT * FROM table WHERE height is NULL/NOTNULL



#LIKE/NOTLIKE

SELECT * FROM table where name LIKE ar%; -- starts with ar

SELECT * FROM table where name LIKE %nn;

SELECT * FROM table where name LIKE %ru%;



#AGG --SUM,MIN,MAX,AVG

SELECT SUM(height) from table;




#ALIASING --AS

SELECT name AS firstname FROM table;




#SORT

SELECT * FROM table ORDER BY name;

SELECT * FROM table ORDER BY name DESC;

SELECT * FROM table ORDER BY name,height;




#GROUP -- WITH AGG

SELECT SUM(height) FROM table GROUP BY weight;



#HAVING

SELECT height FROM table GROUP BY weight HAVING SUM(height) = 120;


#UPDATE


UPDATE table SET height = height + 10 WHERE weight = 50;

update table SET height = 170 WHERE weight = 50;


#DELETE

DELETE FROM table WHERE weight = 70;



#SELF JOIN


SELECT p1.country_code, 
       p1.size AS size2010,
       p2.size AS size2015
FROM populations AS p1
inner  JOIN populations AS p2
ON  p1.country_code = p2.country_code AND p1.size <> p2.size;


#INNER JOIN



SELECT c.code, name, region, e.year, fertility_rate, unemployment_rate
FROM countries AS c
INNER JOIN populations AS p
ON c.code = p.country_code



#LEFT JOIN


SELECT c.code, name, region, e.year, fertility_rate, unemployment_rate
FROM countries AS c
LEFT JOIN populations AS p
ON c.code = p.country_code



#RIGHT JOIN


SELECT c.code, name, region, e.year, fertility_rate, unemployment_rate
FROM countries AS c
RIGHT JOIN populations AS p
ON c.code = p.country_code


#FULL OUTER JOIN


SELECT c.code, name, region, e.year, fertility_rate, unemployment_rate
FROM countries AS c
FULL JOIN populations AS p
ON c.code = p.country_code


#CROSS JOIN ( ALL Possible combination of city and languages)

SELECT c.name AS city, l.name AS language
FROM cities AS c        
CROSS JOIN languages AS l



#INNER JOIN ( 3 TABLES)

SELECT c.code, name, region, e.year, fertility_rate, unemployment_rate
FROM countries AS c
INNER JOIN populations AS p
ON c.code = p.country_code
INNER JOIN economies AS e
on c.code = e.code and e.year = p.year;


#SWITCH WHEN - SELECT INTO:

SELECT country_code, size,
    CASE WHEN size > 50000000 THEN 'large'
        WHEN size > 1000000 THEN 'medium'
        ELSE 'small' END
        AS popsize_group
INTO pop_plus
FROM populations
WHERE year = 2015;

SELECT * FROM pop_plus;


#UNION / UNION ALL ( REPEAT VALUES IF PRESENT IN BOTH TABLES AND ARE STACKED OVER ONE ANOTHER NOT AS NEW COLUMN AS LIKE JOINS )

SELECT country_code
FROM cities
UNION
SELECT code
FROM currencies
ORDER BY country_code;


#INTERSECT ( ALL THE SELECTED VALUES SHOULD MATCH IN BOTH THE TABLES)

SELECT code, year
FROM economies
INTERSECT
SELECT country_code, year
FROM populations
ORDER BY code, year;


#EXCEPT  ( LIKE LEFT JOIN BUT THE MATCHING VALUES ARE NOT INCLUDED IN RESULT)

SELECT name
FROM cities
EXCEPT
SELECT capital
FROM countries
ORDER BY name;


#SEMI JOIN / ANTI JOIN

SELECT DISTINCT name
FROM languages
WHERE code IN / NOT IN
  (SELECT code
   FROM countries
   WHERE region = 'Middle East')
ORDER BY name;


#SUB QUERIES

#WHERE 

SELECT name, country_code, urbanarea_pop
FROM cities
WHERE name IN
  (SELECT capital
   FROM countries)
   
#SELECT

SELECT countries.name AS country,
  (SELECT COUNT(*)
   FROM cities
   WHERE countries.code = cities.country_code) AS cities_num
FROM countries

#FROM 

SELECT local_name, subquery.lang_num
FROM countries,
  (SELECT code, COUNT(*) AS lang_num
   FROM languages
   GROUP BY code) AS subquery
WHERE countries.code = subquery.code


#VIEW

CREATE VIEW Small_country AS
SELECT CustomerName, ContactName
FROM Customers
WHERE Country = "Brazil";

'''

```
